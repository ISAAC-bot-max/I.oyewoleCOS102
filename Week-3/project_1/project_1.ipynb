```python
import tkinter as tk
from tkinter import messagebox, ttk  # Added ttk for the dropdown
import csv

def get_unique_departments():
    """Reads the CSV and returns a sorted list of unique departments."""
    departments = set()
    try:
        with open('employee.csv', mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                departments.add(row['Department'].strip())
    except FileNotFoundError:
        return ["No File Found"]
    
    return sorted(list(departments))

def verify_credentials(u, d):
    try:
        with open('employee.csv', mode='r') as file:
            reader = csv.reader(file)
            next(reader)  # Skip header
            for row in reader:
                if len(row) >= 3:
                    # row[1] is Name, row[2] is Department
                    if row[1].strip().lower() == u.lower() and row[2].strip().lower() == d.lower():
                        return True
    except FileNotFoundError:
        messagebox.showerror("Error", "employee.csv not found.")
    return False

def welcomeMessage(username):
    window = tk.Toplevel(root)
    window.title("Admin Box")
    window.geometry("500x200")
    tk.Label(window, text=f"Welcome {username}\n", font=("Arial", 12, "bold")).pack(pady=10)
    tk.Label(window, text="This is Python GUI with Tkinter").pack()
    tk.Button(window, text="Close", command=window.destroy).pack(pady=20)

def submit():
    user = username_entry.get().strip()
    dept = dept_var.get()  # Get value from the dropdown variable
    
    if not user or dept == "Select Department":
        messagebox.showwarning("Input Error", "Please enter your name and select a department.")
        return

    if verify_credentials(user, dept):
        welcomeMessage(user)
    else:
        messagebox.showinfo("Access Denied", f"Hello {user}, you are not a member of {dept}.")

# Root window setup 
root = tk.Tk()
root.title("Login System")
root.geometry("400x350")

# Name Field
tk.Label(root, text="Full Name:").pack(pady=(20, 0))
username_entry = tk.Entry(root, width=30)
username_entry.pack(pady=5)

# Department Dropdown
tk.Label(root, text="Select Department:").pack(pady=(10, 0))

# Get departments from CSV
dept_options = get_unique_departments()
dept_var = tk.StringVar(root)
dept_var.set("Select Department") # Default value

# Create the Dropdown (OptionMenu)
dept_dropdown = ttk.OptionMenu(root, dept_var, "Select Department", *dept_options)
dept_dropdown.pack(pady=5)

# Login Button
tk.Button(root, text="Login", command=submit, width=15).pack(pady=30)

root.mainloop()
```
