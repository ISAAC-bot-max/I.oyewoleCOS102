```python
import numpy as np


print("----Welcome to Equator----")
print( "1.Cubic Equation : Ax^3 + Bx^2 + Cx + D")
print("2.Quartic Equation : Ax^4 + Bx^3 + Cx^2 + Dx + E")
print("3.Quadratic Equation : Ax^2 + Bx + C")
dec = int(input("Pick an Equation btw(1--3)"))

#code for Cubic 
if dec == 1:
         print("\n Cubic Equation")
         print("Thanks for picking Cubic Equation")
         a = float(input("Enter the values for A: "))
         b = float(input("Enter the values for B: "))
         c = float(input("Enter the values for C: "))
         d = float(input("Enter the values for D: "))
         coefficients = [a, b, c, d]
         roots =np.roots(coefficients)
         print("Roots are:")
         for r in roots:print(r)

#code for Quartic
elif dec == 2:
    print("\nQuartic Equation")
    print("Thanks for picking Quartic Equation")

    a = float(input("Enter value for A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))

    coefficients = [a, b, c, d, e]
    roots = np.roots(coefficients)

    print("Roots are:")
    for r in roots:
        print(r)

#code for quadratic equation
elif dec == 3:
    print("Ax**2 + Bx + C")
    a = float(input("Enter value for a: "))
    b = float(input("Enter  value for b: "))
    c = float(input("Enter value for c: "))

    discriminant = b**2 - 4*a*c

    root1 = (-b + discriminant**0.5) / (2*a)
    root2 = (-b - discriminant**0.5) / (2*a)

    print("Roots are:", root1, root2)
```
