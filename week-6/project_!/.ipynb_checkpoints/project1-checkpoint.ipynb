```python
import tkinter as tk
from tkinter import messagebox
import random

EMPLOYEE_DATABASE = [
    "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
    "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
    "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
    "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
]

TASK_LIST = [
    "Loading", "Transporting", "Reviewing Orders", 
    "Customer Service", "Delivering Items"
]

class Employee:
    def __init__(self):
        self.attendance_log = []
        self.root = tk.Tk()
        self.root.title("Janes Delivery Logistics System")
        self.root.geometry("400x250")
        self.label = tk.Label(self.root, text="Enter Employee Name:")
        self.label.pack(pady=10)
        self.name = tk.Entry(self.root, width=30)
        self.name.pack(pady=5)
        self.verify = tk.Button(
            self.root, text="Verify & Login", command=self.check_employee,
            font=("Arial", 10), fg="blue")
        self.verify.pack(pady=20)

    def check_employee(self):
        user_input = self.name.get().strip()
        if user_input in EMPLOYEE_DATABASE:
            self.take_attendance(user_input)
            task = self.assign_task()
            messagebox.showinfo("Access Granted", 
                                f"Welcome, {user_input}!\n"
                                f"Attendance marked for today.\n"
                                f"Your Task: {task}")
        else:
            self.refuse_access()

    def take_attendance(self, name):
        if name not in self.attendance_log:
            self.attendance_log.append(name)
            print(f"Attendance Log Updated: {self.attendance_log}")

    def assign_task(self):
        task_index = random.randint(0, len(TASK_LIST) - 1)
        return TASK_LIST[task_index]

    def refuse_access(self):
        messagebox.showerror("Access Denied", "Identity verification failed.\n"
                                              "You are not registered in the system.")

app = Employee()
app.root.mainloop()
```
