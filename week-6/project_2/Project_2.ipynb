```python
import tkinter as tk
from tkinter import messagebox


DELIVERY_LOCATIONS = ["PAU", "Epe"]

DELIVERY_RATES = {
    "PAU": {"heavy": 2000, "light": 1500},
    "Epe": {"heavy": 5000, "light": 4000},
}

WEIGHT_THRESHOLD = 10


class DeliveryService:
    def __init__(self):
       
        self.charge_log = []

        # GUI Setup
        self.root = tk.Tk()
        self.root.title("Jane's Delivery Logistics System")
        self.root.geometry("400x250")

       
        self.location_label = tk.Label(self.root, text="Select Delivery Location:", font=("Arial", 12))
        self.location_label.pack(pady=10)

        self.location_var = tk.StringVar(value=DELIVERY_LOCATIONS[0])
        self.location_menu = tk.OptionMenu(self.root, self.location_var, *DELIVERY_LOCATIONS)
        self.location_menu.config(font=("Arial", 12), width=25)
        self.location_menu.pack(pady=5)

        self.weight_label = tk.Label(self.root, text="Enter Package Weight (kg):", font=("Arial", 12))
        self.weight_label.pack(pady=10)

        self.weight_entry = tk.Entry(self.root, font=("Arial", 12), width=30)
        self.weight_entry.pack(pady=5)

        self.calculate_button = tk.Button(
            self.root, text="Calculate Charge", command=self.check_delivery,
            fg="white", font=("Arial", 10, "bold")
        )
        self.calculate_button.pack(pady=20)

    def check_delivery(self):
    
        raw = self.weight_entry.get().strip()

        if not raw:
            messagebox.showwarning("Missing Input", "Please enter the package weight.")
            return

        try:
            weight = float(raw)
            if weight <= 0:
                raise ValueError
        except ValueError:
            messagebox.showerror("Invalid Input", "Please enter a valid positive number for weight.")
            return

        location = self.location_var.get()
        charge = self.get_charge(location, weight)
        self.log_charge(location, weight, charge)

        messagebox.showinfo("Delivery Charge",
                            f"Location: {location}\n"
                            f"Package Weight: {weight}kg\n"
                            f"Amount to Pay: N{charge:,}")

    def get_charge(self, location, weight):
       
        # Reading from the global DELIVERY_RATES variable
        category = "heavy" if weight >= WEIGHT_THRESHOLD else "light"
        return DELIVERY_RATES[location][category]

    def log_charge(self, location, weight, charge):
       
        entry = {"location": location, "weight": weight, "charge": charge}
        self.charge_log.append(entry)
        print(f"Charge Log Updated: {self.charge_log}")

   



    app = DeliveryService()
    app.root.mainloop()
```
