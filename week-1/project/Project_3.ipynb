```python
#Calculating Annuity 
PMT = float(input("Enter Payment amount: "))
R = float(input("Enter Rate (%): ")) / 100
t = float(input("Enter Time (years): "))
n = int(input("Enter number of payments per year: "))

D=(1 + (R/n))**(n*t)
A = PMT * ((D - 1) / (R/n))
print("Future Value of Annuity =", A)
```
