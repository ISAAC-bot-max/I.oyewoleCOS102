```python
#Calculation for Compound Interest 
print("WELCOME")
P = float(input("Enter Principal: "))
R = float(input("Enter Rate(%): "))/100
t = float(input("Enter no..of yeaars: "))
n = int(input("Enter number of times compounded per year: "))
#formula for amount
A = P * (1 + (R/n)) ** (n*t)
CI = A - P 
print("The compound Interest:",CI)

    
```
